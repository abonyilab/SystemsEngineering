# Data Input and Output

This notebook is the reference code for getting input and output, pandas can read a variety of file types using its pd.read_ methods. Let's take a look at the most common data types: 

Nézzük meg mit tud beolvasni és kimenteni a pandas

```python
import numpy as np
import pandas as pd
```

## CSV

### CSV Input

```python
df = pd.read_csv('files\example2.csv')
df
```

### CSV Output

```python
example2 = pd.DataFrame(data=np.random.randn(2,3),index = ['a','b'], columns = ['c1','c2','c3'])
example2
```

```python
example2.to_csv('files\example2.csv',index=False)
```

## Excel
Pandas can read and write excel files, keep in mind, this only imports data. Not formulas or images, having images or macros may cause this read_excel method to crash. 

### Excel Input

```python
pd.read_excel('files\Excel_Sample.xlsx',sheet_name='Sheet1')
```

### Excel Output

```python
df.to_excel('files\Excel_Sample.xlsx',sheet_name='Sheet1')
```
