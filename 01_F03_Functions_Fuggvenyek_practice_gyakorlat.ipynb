# Volume of a sphere
Write a function that computes the volume of a sphere given its radius. / Írj egy függvényt, amit kiszámítja egy gömb térfogatát a sugarának ismeretében!


The volume of a sphere is given as / A gömb térfogata a következő egyenlettel számítható $$\frac{4}{3} πr^3$$

```python
def vol(rad):
    return 4/3*3.14*rad**3
```

```python
# Check
vol(2)
```

# Product of list elements
Write a Python function to multiply all the numbers in a list. / Írj egy Python függvényt, ami összeszorozza egy lista elemeit.

    Sample List : [1, 2, 3, -4]
    Expected Output : -24

```python
def multiply(numbers):  
    pr = numbers[0]
    for i in numbers[1:]:
        pr = pr*i
    print(pr)
```

```python
multiply([1,2,3,-4])
```

Hozzunk létre egy függvényt ami az $x^2-2$ függvény értékét számítja ki tetszőleges x bemenet esetén.

```python
def my_f_func(x):
    return x**2-2
```

```python
fgv_x = [x*0.1 for x in range(-40,40,1)]
fgv_ert = [my_f_func(x*0.1) for x in range(-40,40,1)]

import matplotlib.pyplot as plt
plt.plot(fgv_x,fgv_ert)
plt.show()
```
