## Basic Matplotlib Commands / Az alapvető séma 


```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
plt.plot(x, y, 'r') # 'r' is the color red / színek beállítsa ... 
plt.xlabel('X Axis Title Here') #tengelyek feliratai 
plt.ylabel('Y Axis Title Here') 
plt.title('String Title Here') #ábrák címe 
plt.show()
```

## Creating Multiplots on Same Canvas / Osztott ábrák

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')  #a jelölések mint matlab-ban 
plt.subplot(1,2,2)
plt.plot(y, x, color="blue", lw=3, linestyle='-', alpha=0.1, marker='x') # alpha=0.5 half-transparant' - áttetsző 
#lw: linewidth - vastagság 
```

## Introduction to the Object Oriented Method 

## Objektumorientált megközelítésmód 

Create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

Picit macerás, de majd kifizetődő technika, hogy elsőként fig objektumot hozunk léte, majd ehhez kapcsolódó metódust hívunk meg vagy tulajdonságot állítunk ...

add_ - ábrához adunk ...

set_ - tulajdonságot állítunk 



```python
fig = plt.figure() # Create Figure (empty canvas) / új ábra mint objektum - egy üres vászon 

# Add set of axes to figure / hova kerüljön a vásznon az ábra 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'b', label="x**2") #"belerajzolunk"
axes.plot(x, x**3, 'r', label="x**3") #"belerajzolunk"

axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods / most onbjektum tulajdonságát állítjuk 
axes.set_ylabel('Set y Label') # set_ metódus hívás!!!! 
axes.set_title('Set Title')
axes.legend()  #cimkék ..
axes.legend(loc=2) # upper right corner - melyik sarokba (1-4)
```

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

Most jön mi a kompikáltság előnye, a vásznon teszőlegesen pozicionálhatjuk az ábráinkat 

```python
fig = plt.figure()

abra1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes / fő-ábra a sarkok potivíóival

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes /al-ábra a pozicíóval 

abra1.plot(x, y, 'b')
abra1.set_xlabel('X_label_axes2')
abra1.set_ylabel('Y_label_axes2')
abra1.set_title('Axes 2 Title')

axes2.plot(x, y, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');


axes2.set_yscale("log") # logarimikus tengely
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Ua mint előbb, csak subplotokat generálunk


```python
cimek=['elso','masodik']
fig, axes = plt.subplots(nrows=1, ncols=2) # létrehozzuk a canvas-t és a koordináta rendszereket
k=0
for ax in axes: # ciklusban végigmegyünk a koordinátákon
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(cimek[k])
    k+=1
    ax.set_ylim([0, 40]) #sez the y and x axis ...  tengelyek beállítása
    ax.set_xlim([2, 5])

fig.tight_layout() #fig.tight_layout() or plt.tight_layout() automatically adjusts - igazítás, probald ki nelkule


```

### Figure size, aspect ratio and DPI

### Ábrák mérete, aránya, felbontása

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

Az ábrák létrehozásakor a `figsize` and `dpi` bemenetekkel határozhatjuk meg az ábrák méretét és felbontását. 

```python
fig, axes = plt.subplots(figsize=(12,3), dpi=100)

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```
